```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
wiki='https://en.wikipedia.org/wiki/Main_Page'
result=requests.get(wiki)
result
```

```python
content=result.text
soup=BeautifulSoup(content, 'lxml')
soup
```

```python
contents=soup.find('div',{'id': 'mp-upper'})
contents
```

```python
import re
links=contents.findAll('a', href=re.compile(r"(/wiki/)+([A-Za-z0-9_:()]+)"))
len(links)
```

```python
title=[]
all_links=[]
for link in links:
    if len(link)==0:
        title.append('No Title')
    else:
        title.append(link.text)
    all_links.append(link['href'])
all_links
```

```python
n=range(1, len(all_links)+1)
ind=pd.Index(n)
dataset=pd.DataFrame({'Title': title, 'Link':all_links}, index=ind)
dataset
```

```python
dataset.to_csv('Wikipedia scrap')
```
