```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
web='https://www.flipkart.com/search?sid=czl&otracker=CLP_Filters&p%5B%5D=facets.brand%255B%255D%3DSAMSUNG'
response = requests.get(web, headers={'User-agent': 'Super Bot Power Level Over 9000'})
response
```

```python
content=response.text
soup=BeautifulSoup(content, 'lxml')
soup
```

```python
containers=soup.findAll('div', {'class': '_1YokD2 _3Mn1Gg'})
containers
```

```python
contents=containers[0].findAll('div', {'class': '_4ddWXP'})
len(contents)
```

```python
container=contents[0]
product_name=container.findAll('a', {'class': 's1Q9rs'})
product_name[0].text
```

```python
products= []
discount_price =[]
original_price =[]
percent_off=[]
category=[]
for contain in contents:
    product_name = contain.findAll('a', {'class': 's1Q9rs'})
    if len(product_name)==0:
        products.append(np.nan)
    else:
        name=product_name[0].text
        products.append(name)
    cat= contain.findAll('div', {'class': '_3Djpdu'})
    if len(cat)==0:
        category.append(np.nan)
    else:
        cate=cat[0].text
        category.append(cate)
    orig_price=contain.findAll('div',{'class': '_3I9_wc'})
    if len(orig_price)==0:
        original_price.append(np.nan)
    else:
        ori_price=orig_price[0].text
        original_price.append(ori_price)
    disc_price=contain.findAll('div', {'class': '_30jeq3'})
    if len(disc_price)==0:
        discount_price.append(np.nan)
    else:
        dis=disc_price[0].text
        discount_price.append(dis)
    perc=contain.findAll('div', {'class': '_3Ay6Sb'})
    if len(perc)==0:
        percent_off.append(np.nan)
    else:
        percentage=perc[0].text
        percent_off.append(percentage)
original_price
```

```python
new=pd.DataFrame({'Product Name': products, 'Category': category, 'Orignal Price ': original_price,
                  'Discount Price ': discount_price, 'Percentage_Off %': percent_off})
new
```

```python
new.to_csv('Television')
```
